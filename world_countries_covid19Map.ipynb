```python
def lw(i,color):
    folium.Marker(location=point,
    popup=str([dataset['Cases'][i],dataset['Deaths'][i]])[1:-1],
    icon=folium.Icon(color=color,icon='tint',icon_color="white")
                 ).add_to(cmap)
```

```python
i=0
for point in place:
    if dataset['Cases'][i]<'50000':
        lw(i,'green')
    if dataset['Cases'][i]>'50000' and dataset['Cases'][i]<'1000000':
        lw(i,'orange')
    elif dataset['Cases'][i]>'100000' and  dataset['Cases'][i]<'300000':
        lw(i,'red')
    elif dataset['Cases'][i]>'30000000':
        lw(i,'darkred')
    i+=1
```
